## Create a Simple Time Series using Python

$$
y_t = \alpha+\beta y_{t-1}+ \xi_{t}.
$$

上面的生成序列告诉我们今天的观测值($y_t$)是以常数α和昨天的观测值βyt-1的比例计算的，再加上一个随机的error$\xi_t$。

```python
import pandas  as pd
import numpy  as np
seed = 2016
np.random.seed(seed)
y_0 = 1
alpha = -0.25
beta = 0.95
y = pd.Series(y_0)
num = 10
for i in range(num):
    y_t = alpha + (beta*y_0) + np.random.uniform(-1, 1)
    y.at[i] = y_t
    y_0 = y_t
# 前两行导入pandas和numpy库。这两个库是数据科学的双主力。
# 系列的初始值(y0)存储在Python变量y_0中。
# 方程参数α,β分别由α和β来捕获，随机误差()由调用random.uniformmethod产生，并被约束在-1和+1之间。
# 随机误差()通过调用random.uniform方法生成，并被约束在-1和+1之间。
# 代码通过设置(num=10)并使用一个简单的for循环生成10个观测值。
```
