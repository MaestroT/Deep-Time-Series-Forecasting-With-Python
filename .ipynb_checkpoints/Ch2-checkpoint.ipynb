## Generating Data for use with a Deep Neural Network
我们将建立一个DNN来逼近$y=x^2$。

```python
import pandas as pd
import numpy as np
import random
```

```python
random.seed(2020)
sample_size = 50
sample = pd.Series(random.sample(range(-10000,10000), sample_size))
# random.sample函数用于生成50个观测值的随机样本，其结果存储为一个数据类型，称为Series。这是通过参数pd.Series()实现的。
# 一个Series是一个一维的标签数组。 它可以包含任何数据类型 (整数、字符串、浮点数、Python 对象等)。
x = sample / 10000
y = x**2
```

## Translating Sample Data into a Suitable Format
在这个例子中，将使用neuralpy包。它是一个专门研究神经网络的机器学习库。它直观的界面让你可以轻松地快速开始训练数据和测试深度学习模型。为了使用这个包，我们需要把数据放到一个合适的格式中。要做到这一点，我们将创建一个名为dataSet的Python对象，它将包含x和y观测值的列表。一个简单的方法是通过一个 while 循环来实现的

```python
count = 0
dataSet = [([x.iloc[count]],[y.iloc[count]])]
count = 1
while(count < sample_size):
    print("Working on data item:{}".format(count))
    dataSet = (dataSet +[([x.loc[count]],[y.iloc[count]])])
    count = count + 1
```
